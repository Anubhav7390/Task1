```python
df1 = df.dropna(subset = ['Income'], ignore_index = True)   # Droping rows with missing income (only 24 out of 2240)
df1
```

```python
# adding age column

from datetime import datetime

df1["Age"] = datetime.now().year - df1['Year_Birth']

df1
```

```python
# total children

df1["Total_Children"] = df['Kidhome'] + df['Teenhome']
df1
```

```python
# convertion to datetime of column 'Dt_Customer'

df1['Dt_Customer'] = pd.to_datetime(df1['Dt_Customer'], dayfirst=True)

df1
```

```python
# finding tenure of customers

df1["Customer_Tenure_Days"] = (pd.to_datetime("today") - df1["Dt_Customer"]).dt.days

df1
```

```python
# Total Spending

spent_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']  

df1["Total_spendings"] = df1[spent_cols].sum(axis=1)
df1
```

```python
#  Segmentation of Income column

def income_segment(income):
    if income< 30000:
        return "Low"
    elif income < 60000:
        return "Mid-Low"
    elif income < 90000:
        return "Mid"
    elif income < 120000:
        return "Mid-High"
    else:
        return "High"

df1["Income_segment"] = df["Income"].apply(income_segment)
df1
```
